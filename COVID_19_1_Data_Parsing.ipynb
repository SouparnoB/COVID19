## Extracting Population and Area of Provinces in Different Countries

### Provinces of Spain

```python
src=requests.get('https://en.wikipedia.org/wiki/Autonomous_communities_of_Spain').text
```

```python
soup=BeautifulSoup(src,'html.parser')
```

```python
#provinces of spain:
province=[]
area=[]
pop=[]
for i in enumerate(soup.find_all('table',class_='sortable wikitable')):
    a=i[0]
    b=i[1]
    for m in enumerate(b.find_all('td')):
        #print(a)
        #print(m)
        #n=m[0]
        #d=m[1].text
        #print(d)
        #print(n,d)
        #print('Next:\n')
        if a==0:
            if m[0]%12==1:
                b=m[1].text
                b=b[:-1]
                province.append(b)
            elif m[0]%12==7:
                b=m[1].text
                b=b[:-1]
                area.append(b)
            elif m[0]%12==8:
                b=m[1].text
                b=b[:-1]
                pop.append(b)
        elif a==1:
            if m[0]%11==2:
                b=m[1].text
                b=b[:-1]
                province.append(b)
            elif m[0]%12==7:
                b=m[1].text
                b=b[:-1]
                area.append(b)
            elif m[0]%12==8:
                b=m[1].text
                b=b[:-1]
                pop.append(b)
```

```python
spain_demo_data={'province_name':province,'area_km2':area,'population':pop}
spain_demo_data=pd.DataFrame(spain_demo_data)
spain_demo_data.dropna()
spain_demo_data['country']='Spain'
spain_demo_data.head()
```

```python
#provinces of italy:
province=[]
city=[]
area=[]
pop=[]
for i in enumerate(soup.find_all('table',class_='wikitable sortable mw-datatable')):
    #print(i)
    a=i[0] 
    b=i[1]
    #a
    for m in enumerate(b.find_all('td')):
        #print(a)
        #print(m)
        n=m[0]
        d=m[1].text
        #print(d)
        #print(n,d)
        #print('Next:\n')
        if m[0]%11==1:
                b=m[1].text
                b=b[:-1]
                city.append(b)
        elif m[0]%11==4:
            b=m[1].text
            b=b[:-1]
            province.append(b)
        elif m[0]%11==6:
            b=m[1].text
            b=b[:-1]
            pop.append(b)
        elif m[0]%11==7:
            b=m[1].text
            b=b[:-1]
            area.append(b)
```
